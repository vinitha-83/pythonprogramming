```python
# Finding the column with the highest number of missing values in the data_1 DataFrame
# Printing the column name with the highest number of missing values


nul_data_1=data_1.isnull().sum().idxmax()
print(nul_data_1)
```

```python
#Grouping the data by the 'date' column and then calculating the sum of the 'personnel' column for each unique date using the sum() function. 

daily_death=data_2.groupby('date')['personnel'].sum()
print(daily_death)
```

```python
#creating  a histogram using matplotlib to visualize the daily death toll in the data_set_2

plt.hist(daily_death,bins=5,edgecolor='black',facecolor='green')

#setting the labels for x and y axes and title to the plot

plt.xlabel('Daily Death Toll')
plt.ylabel('Frequency')
plt.title('Daily Death Toll',fontsize=20,color='red')
plt.show()
```
